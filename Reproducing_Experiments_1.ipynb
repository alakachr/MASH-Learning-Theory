```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import Ridge
```

The inputs are sampled from the uniform distribution in [âˆ’1, 1],
while the optimal regression function is a degree 2 polynomial (blue curve below). Gaussian noise is added
to generate the outputs (black crosses below).

```python
k = 5
n = 10
x = np.sort(np.random.uniform( -1,1, n))
#x = np.array(range(n))
eps = np.random.normal(0,0.3,size= (n, 1))

phi = np.ones((n, k+1))
for i in range(k):
    phi[ :, i+1] = x**(i+1)

theta  = np.array([[-0.5, 0, 1, 0,0,0]])
y = np.dot(phi, theta.T) + eps


#test
n_test=50
x_test = np.linspace(-1, 1, num=50)
phi_test = np.ones((n_test, k+1))
for i in range(k):
    phi_test[ :, i+1] = x_test**(i+1)
y_curve= x_test**2-0.5 #optimal predicor

eps = np.random.normal(0,0.3,size= (n_test, 1)) #refresh noise
y_test = np.dot(phi_test, theta.T) + eps

reg = linear_model.LinearRegression()
reg.fit(phi, y)

y_pred = reg.predict(phi_test)

#compute mse
train_mse = mean_squared_error(y,reg.predict(phi))
test_mse = mean_squared_error(y_test,reg.predict(phi_test))
```

```python
plt.plot(x_test, y_pred, "red")
plt.plot(x, y, "x")
plt.plot(x_test, y_curve, "blue")
plt.title('n='+str(n)+" " + "train="+str(round(train_mse, 5))+" " + "test= "+str(round(test_mse, 4)))
plt.show()

print(train_mse, test_mse)
```

```python
### Risk bound

#random design

#sigma =  E[phi(x), phit(x).T]
#      = [(E0.5Â²)=0.25     EXÂ²0.5=1/6
#        EXÂ²0.5 =1/6        E(XÂ²)Â² =1/5]
    
    
    
Sigma = np.array([[0.25,        float(1/6)] , 
                  [float(1/6),   float(1/5)]])

theta_star  = np.array([[-0.5, 1]])


k = 2
n = 100
x = np.sort(np.random.uniform( -1,1, n))
#x = np.array(range(n))
eps = np.random.normal(0,0.3,size= (n, 1))

phi = np.ones((n, k))

phi[ :, 1] = x**(2)


y = np.dot(phi, theta_star.T) + eps

reg = linear_model.LinearRegression()
reg.fit(phi, y)



theta_hat = np.array([reg.intercept_[0], reg.coef_[0][1] ])
print(theta_hat, )
plt.plot(x, y, "x")

diff = theta_hat - theta_star
excess_risk = np.dot(Sigma, diff.T)
excess_risk=np.dot(diff,excess_risk )
print(excess_risk)
bound =(0.3**2)*k/n
print(bound)



```

```python
list_n = np.array([10,50,100,500,1000,2000,5000, 6000,7000, 10000, 20000, 50000, 100000])
k=2
list_excess_risk = list()
list_train_risk = list()
for n in list_n:
    for i in range(500):
        
        sublist_excess_risk= list()
        sublist_train_risk= list()
        
        theta_star  = np.array([[-0.5, 1]])


        k = 2
        
        x = np.sort(np.random.uniform( -1,1, n))
        #x = np.array(range(n))
        eps = np.random.normal(0,0.3,size= (n, 1))

        phi = np.ones((n, k))

        phi[ :, 1] = x**(2)
        
        Sigma = np.dot(phi.T, phi)/n

        y = np.dot(phi, theta_star.T) + eps

        reg = linear_model.LinearRegression()
        reg.fit(phi, y)

        theta_hat = np.array([reg.intercept_[0], reg.coef_[0][1] ])
        
        

        diff = theta_hat - theta_star
        excess_risk = np.dot(diff,Sigma)
        excess_risk=np.dot(excess_risk,diff.T )


        sublist_excess_risk.append(excess_risk)

        
    list_excess_risk.append(np.mean(sublist_excess_risk))
    
```

```python
plt.plot(np.log10(list_n), np.log10(list_excess_risk))
plt.plot(np.log10(list_n), np.log10((0.3**2)*2/list_n), "red", )
plt.legend(("excess risk", "bound"))
plt.show()
```

```python
list_n = np.array([10,50,100,500,1000,2000,5000, 6000,7000, 10000, 20000, 50000, 100000])
k=5
list_excess_risk = list()
list_train_risk = list()
for n in list_n:
    for i in range(100):
        
        sublist_excess_risk= list()
        sublist_train_risk= list()
        
        x = np.sort(np.random.uniform( -1,1, n))
        #x = np.array(range(n))
        eps = np.random.normal(0,0.3,size= (n, 1))

        phi = np.ones((n, k+1))
        for i in range(k):
            phi[ :, i+1] = x**(i+1)

        theta  = np.array([[-0.5, 0, 1, 0,0,0]])
        y = np.dot(phi, theta.T) + eps


        #test
        n_test=1000
        x_test = x = np.sort(np.random.uniform( -1,1, n_test))
        phi_test = np.ones((n_test, k+1))
        for i in range(k):
            phi_test[ :, i+1] = x_test**(i+1)
        
        eps = np.random.normal(0,0.3,size= (n_test, 1)) #refresh noise
        y_test = np.dot(phi_test, theta.T) + eps

        reg = linear_model.LinearRegression()
        reg.fit(phi, y)

        y_pred = reg.predict(phi_test)

        #compute mse
        train_mse = mean_squared_error(y,reg.predict(phi))
        test_mse = np.abs(mean_squared_error(y_test,y_pred) -np.var(eps))
        
        

        sublist_excess_risk.append(test_mse)
        sublist_train_risk.append(train_mse)
        
    list_excess_risk.append(np.mean(sublist_excess_risk))
    list_train_risk.append(np.mean(sublist_train_risk))
    
    

   
```

```python
plt.plot(np.log10(list_n), np.log10(list_excess_risk))
plt.plot(np.log10(list_n), np.log10((0.3**2)*+(k+1)/list_n), "red", )
plt.legend(("excess risk", "bound"))
plt.show()
```

```python
######## Ridge #######



list_lambda = np.linspace(0.000001, 1000, 10000)
k=5
n= 20
list_excess_risk = list()
list_bias = list()
list_var = list()
for l   in list_lambda:
    for i in range(32):
        
        sublist_excess_risk= list()
        
        np.random.seed(seed=1)
        x = np.sort(np.random.uniform( -1,1, n))
        np.random.seed()
            
        #x = np.array(range(n))
        eps = np.random.normal(0,0.1,size= (n, 1))

        phi = np.ones((n, k+1))
        for i in range(k):
            phi[ :, i+1] = x**(i+1)

        theta  = np.array([[0, 0, 1, 0,0,0]])
        y = np.dot(phi, theta.T) + eps


        reg = Ridge(l)
        reg.fit(phi, y)

        
        theta_hat = np.array([reg.intercept_[0], reg.coef_[0][1], reg.coef_[0][2], reg.coef_[0][3] , reg.coef_[0][4], reg.coef_[0][5]])
        
        Sigma = np.dot(phi.T, phi)/n

        diff = theta_hat - theta
        excess_risk = np.dot(diff,Sigma)
        excess_risk=np.dot(excess_risk,diff.T )


        sublist_excess_risk.append(np.reshape(excess_risk, (1,)))
        
    list_excess_risk.append(np.mean(sublist_excess_risk))
    
    temp = np.dot(np.linalg.matrix_power(Sigma + l*np.identity(6), -2), Sigma)
    bias = l*l*(np.dot(np.dot(theta,temp), theta.T))
    
    print( bias)
    list_bias.append(np.reshape(bias, (1,)))
    
    temp = np.dot(Sigma, np.linalg.inv(Sigma + l*np.identity(6)))
    variance = 0.01* np.trace(np.dot(temp, temp))/n
    list_var.append(np.reshape(variance, (1,)))
    
    
```

```python
plt.plot(np.log10(list_lambda),np.log10(list_excess_risk))
plt.plot(np.log10(list_lambda), np.log10(list_var))
plt.plot(np.log10(list_lambda), np.log10(list_bias))
plt.plot(np.log10(list_lambda), np.log10(np.array(list_var) + np.array(list_bias)))
plt.legend(("excess risk", "var", "bias", "bias +variance"))
```
